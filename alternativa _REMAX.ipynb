```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re

options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')

# driver_path = 'C:\\Users\\LENOVO\\Downloads\\chromedriver_win32\\chromedriver.exe'
driver_path = "./chromedriver_win32/chromedriver.exe"


webD = webdriver.Chrome(driver_path)

webD.get("https://remax.pe")

# inicializarla en la pantalla 2:

webD.set_window_position(2000,0)
webD.maximize_window()
sleep(3)
#selecciono tipo
webD.find_element_by_xpath('//*[@id="btn_buscar"]').click()
sleep(3)
# marco terreno urbano
webD.find_element_by_xpath('//*[@id="combopropiedad"]').click()
sleep(3)

#selecciono departamento
webD.find_element_by_xpath('/html/body/section[3]/div/div/div[1]/div/div[2]/ul/li[2]/div/select/option[18]').click()
sleep(3)
# marco lima
webD.find_element_by_xpath('//*[@id="combodepartamento"]').click()
sleep(3)

#selecciono provincia
webD.find_element_by_xpath('//*[@id="combodepartamento"]/option[17]').click()
sleep(3)
# marco lima
webD.find_element_by_xpath('//*[@id="comboprovincia"]').click()
sleep(3)

#selecciono provincia
webD.find_element_by_xpath('//*[@id="comboprovincia"]/option[11]').click()
sleep(3)
# marco lima
webD.find_element_by_xpath('//*[@id="divdistrito"]/span').click()
sleep(3)

#selecciono distrito
webD.find_element_by_xpath('//*[@id="divdistrito"]/span/div/ul/li[32]/a/label').click()
sleep(3)
# marco san juan de lurigancho
webD.find_element_by_xpath('//*[@id="btenviar"]').click()
sleep(3)

# elaboramos un contador de anuncios para obtener el rango del bucle
terrenos = webD.find_elements_by_xpath('//div[@class="cards__content cards__content--right col-lg-7 col-md-12"]')
# oooo = webD.find_elements_by_xpath('//div[@class="area"]/ul/li[2]/a')

# listas vacias preparadas para la data
precio = []

distrito = []

area = []

ddetallada = []

indice = []

n = 0


for i in range(1,len(terrenos)+1,1):
    
    
    webD.find_element_by_xpath("//div/div/div[2]/div[2]/div[%s]/div/div[2]/div[1]/div[5]/a"%(i)).click()
    print(i)
    sleep(3) 
    
    inf1 = webD.find_element_by_xpath("//*[@id='propiedadDetalle']/div/div/div[1]/div[1]/h2").text
    precio.append(inf1)
    
    inf2 = webD.find_element_by_xpath("//*[@id='propiedadDetalle']/div/div/div[1]/div[1]/ul[2]/li").text
    distrito.append(inf2)
    
    inf3 = webD.find_element_by_xpath("//*[@id='propiedadDetalle']/div/div/div[1]/div[2]/div/div[1]/div/div/h4").text
    area.append(inf3)
    
    inf4 = webD.find_element_by_xpath("//*[@id='propiedadDetalle']/div/div/div[1]/div[3]/div/div[2]/div").text
    ddetallada.append(inf4)
    
    n = n + 1
    indice.append(n)

    webD.back()
    sleep(3)
```

```python
datos = {"PRECIO" : precio, "DISTRITO" : distrito , "AREA" : area , "DESCRIPCIÓN" : ddetallada}

registro = pd.DataFrame(datos,index = indice )
# registro.to_excel( 'C:\\Users\\LENOVO\\Desktop\\DATA_REMAX.xlsx')
registro.to_excel( '../../../Desktop/DATA_CAS.xlsx')
```
