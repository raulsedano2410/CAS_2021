```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By


import time
from time import sleep
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import re

options = webdriver.ChromeOptions()
options.add_argument('--start-maximized')
options.add_argument('--disable-extensions')


# Ruta absoluta:

# driver_path = 'C:\\Users\\LENOVO\\Downloads\\chromedriver_win32\\chromedriver.exe'

#Ruta relativa

driver_path = "./chromedriver_win32/chromedriver.exe"


driver = webdriver.Chrome(driver_path, chrome_options=options)

# Inicializamos el navegador
driver.get('https://www.convocatoriascas.com')


# inicializarla en la pantalla 2:

driver.set_window_position(2000,0)
driver.maximize_window()

#creando listas vacias

institucion =[]
tipo_de_contrato = []
n_vacantes = []
fecha_limite = []
lugar_de_trabajo = []
remuneracion = []
puestos_disponibles_para = []
formacion_academica = []

indice = []
j=0


# esta linea nos ayudara a retener los titulos y darnos la cantidad de elementos con el metodo len:

convocatorias = driver.find_elements_by_xpath('//article[@class="post"]')


# aca empesamos el nuevo procedimiento de entrar y salir dando clicks de todos los anuncios

for i in range(1,len(convocatorias)+1,1):
    
# sin este try-except el  WebDriverWait no corre

    try:

        driver.find_element_by_xpath("//article[%s][@class='post']//a[contains(text(),'')]"%(i)).click()
        print(i)
        #sleep(3)
        
        # Institucion
        inf1 = driver.find_element_by_xpath("//div[@class='conv-item datos']/p[1]").text
        #print(i)
        print("codigo jala data")
        
        if inf1:
            uno = inf1.split("\n")
            dos=uno[0].split(":")
            tres = dos[1]

# Tipo de Contrato
        inf2 = driver.find_element_by_xpath("//div[@class='conv-item datos']/p[2]").text
        if inf2 :
            inf2_1=inf2.split(":")
            inf2_2=inf2_1[1]
        
# Numero de vacantes 
        inf3 = driver.find_element_by_xpath("//div[@class='conv-item datos']/p[3]").text
        if inf3 :
            inf3_1=inf3.split(":")
            inf3_2=inf3_1[1]
        
#Fecha Limite
        inf4 = driver.find_element_by_xpath("//div[@class='conv-item datos']/p[4]").text
        if inf4 :
            inf4_1=inf4.split(":")
            inf4_2=inf4_1[1]

# Lugar de tabajo
        inf5 = driver.find_element_by_xpath("//div[@class='conv-item datos']/p[5]").text
        if inf5:
            inf5_1=inf5.split(":")
            inf5_2=inf5_1[1]
        

# Remuneracion
        inf6 = driver.find_element_by_xpath("//div[@class='conv-item datos']/p[6]").text
        if inf6:
            inf6_1=inf6.split(":")
            inf6_2=inf6_1[1]
        
# Puestos Disponibles
        inf7 = driver.find_element_by_xpath("//div[@class='conv-item datos']/p[7]").text
        if inf7:
            inf7_1=inf7.split(":")
            inf7_2=inf7_1[1]
        
#formacion academica        
        inf8 = driver.find_element_by_xpath("//div[@class='conv-texto1']/p[1]").text
        if inf8:
            inf8_1=inf8.split(":")
            inf8_2=inf8_1[1]

    except:
        #print(i)
        print("codigo no jala data")
        tres = "no info"
        inf2_2 = "no info"
        inf3_2 = "no info"
        inf4_2 = "no info"
        inf5_2 = "no info"
        inf6_2 = "no info"
        inf7_2 = "no info"
        inf8_2 = "no info"
        
    finally:
        # Append data
        institucion.append(tres)
        tipo_de_contrato.append( inf2_2 )
        n_vacantes.append(inf3_2)
        fecha_limite.append(inf4_2)
        lugar_de_trabajo.append(inf5_2)
        remuneracion.append(inf6_2)
        puestos_disponibles_para.append(inf7_2)
        formacion_academica.append(inf8_2)
        
        # para el indice
        j = j + 1
    
        indice.append(j)
                
        
    driver.back()
    sleep(3)
    
    
    datos = {"Institucion" : institucion, "Tipo de Contrato" : tipo_de_contrato , "Numero de Vacantes" : n_vacantes  , "Fecha_Limite" : fecha_limite,"Lugar_de_trabajo" : lugar_de_trabajo,"Sueldo":remuneracion,"Puestos Disponibles" :puestos_disponibles_para,"Formacion_academica":formacion_academica }
```

```python
#creamos el Dataframe
registro = pd.DataFrame(datos,index = indice)

registro.to_excel( '../../../Desktop/DATA_CAS.xlsx')
```
